### Data Cleaning 

```python
#lets count the missing percentage of the columns that contains the null values 

missing_percentage = (data.isnull().sum()/len(data)) * 100 
missing_percentage.sort_values(ascending = False)
```

```python
# drop the columns which have more than 90% missing values 

cols_to_drop = missing_percentage[missing_percentage >= 90].index
drop_data = data.drop(columns = cols_to_drop)
drop_data
```

```python
drop_data.columns
```

```python
data1 = drop_data.drop(columns = ['job_id','company_id','sponsored','posting_domain','zip_code', 'fips','currency','job_posting_url','application_url'])
data1
```

```python
data1.columns
```

```python
data1.isnull().sum().sort_values(ascending=False).head(11)
```

```python
# replace null values with "not mentioned" for accurate analysis 

data1["remote_allowed"] = data1["remote_allowed"].fillna("Not Mentioned")
data1["compensation_type"] = data1["compensation_type"].fillna("compensation_type")
data1["pay_period"] = data1["pay_period"].fillna("pay_period")
data1["formatted_experience_level"] = data1["formatted_experience_level"].fillna("formatted_experience_level")
data1["company_name"] = data1["company_name"].fillna("company_name")
data1["description"] = data1["description"].fillna("description")

avg_views = data1["views"].mean()
data1["views"] = data1["views"].fillna("avg_views")

```

```python
data1.isnull().sum().sort_values(ascending=False).head(11)
```

```python
#  Fill missing values using the mean for each job title group mean 

data1['applies'] = data1['applies'].fillna(data1.groupby('title')['applies'].transform('mean'))
data1['max_salary'] = data1['max_salary'].fillna(data1.groupby('title')['max_salary'].transform('mean'))
data1['min_salary'] = data1['min_salary'].fillna(data1.groupby('title')['min_salary'].transform('mean'))
data1['normalized_salary'] = data1['normalized_salary'].fillna(data1.groupby('title')['normalized_salary'].transform('mean'))
```

```python
# Fill any remaining missing values using the overall dataset mean
#Group mean → average within each category 
#Overall mean → average for the whole dataset


data1['applies'] = data1['applies'].fillna(data1['applies'].mean())
data1['max_salary'] = data1['max_salary'].fillna(data1['max_salary'].mean())
data1['min_salary'] = data1['min_salary'].fillna(data1['min_salary'].mean())
data1['normalized_salary'] = data1['normalized_salary'].fillna(data1['normalized_salary'].mean())
```

### Word Frequency Count 

```python
# let's count the words in title column so we get the common word in a job position
from collections import Counter                               # load the counter library

from nltk.corpus import stopwords                              # load the stopwords library
import nltk
nltk.download('stopwords')                                       # download stopwords (only needed once)
stop_words = set(stopwords.words('english'))                     # load English stopwords
all_title_words = []                                      # create a blank list to store split words in it
for title in data1['title']:                                  # loop for split the words and store it in a blank list
 words = title.split()
 filtered_words = [word for word in words if word.lower() not in stop_words]         # remove stopwords
 all_title_words.extend(filtered_words)                                         # add filtered words to the list
title_word_freq = Counter(all_title_words)                                              # count the split words
print(title_word_freq.most_common(10))                                    # print the 10 most common words
```

```python
# let's count the words in description column so we get the common word in a job description

from nltk.corpus import stopwords # import stopwords
import nltk
stop_words = set(stopwords.words('english')) 
all_description_words = [] 
for description in data1['description']:
    words = description.split()
    filtered_words = [word for word in words if word.lower() not in stop_words] 
    all_description_words.extend(filtered_words) 

    
description_word_freq = Counter(all_description_words)

print(description_word_freq.most_common(25)) 
```

```python
# let's count the words in location column so we get the common location in a job position

from collections import Counter 
from nltk.corpus import stopwords 
import nltk
nltk.download('stopwords') 
stop_words = set(stopwords.words('english')) 
all_location_words = []
for location in data1['location']: 
    words = location.split()
    filtered_words = [word for word in words if word.lower() not in stop_words] 
    all_location_words.extend(filtered_words)

    
location_word_freq = Counter(all_location_words) 
print(location_word_freq.most_common(10)) 
```

```python
# create a category function for views columns

def views_cat_func(x): 
    if(x == "Not-Mentioned"):
        return "Not-Mentioned"
    elif(x<100):
        return "Low-Views"
    elif(x>100 and x<=250):
        return "Medium-Views"
    elif(x>250 and 500):
        return "High-Views"
    else:
        return "Excellent-Views"
        
data1 ['views_cat_func'] = data1['views'].apply(views_cat_func)
```

```python
# print the views category column & count the values

data1.views_cat_func.value_counts()
```

```python
# Calculate counts
views_counts = data1['views_cat_func'].value_counts().sort_index()

# Line plot
sns.lineplot(x=views_counts.index, y=views_counts.values, marker='o', color='green')
plt.title('Job Posting views_cat_func')
plt.xlabel('Views Category')
plt.ylabel('Count the Views Category')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# create a category function for applies column

def applies_cat_func(x):
    if(x == "Not-Mentioned"):
        return "Not-Mentioned"
    elif(x<10):
        return "Low-Applies"
    elif(x>10 and x<=25):
        return "Medium-Applies"
    elif(x>25 and 50):
        return "High-Applies"
    else:
        return "Very-High-Applies"
    
data1['applies_cat_func'] = data1['applies'].apply(applies_cat_func)
```

```python
# Calculate counts
applies_counts = data1['applies_cat_func'].value_counts()

# Pie chart

plt.figure(figsize=(5,5))
plt.pie(applies_counts.values, labels=applies_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'),textprops={'fontsize': 10} )
plt.title('Job Posting applies_cat_func Category', fontsize=16, fontweight='bold')
plt.axis('equal') # Circle shape maintain karva
plt.show()
```
